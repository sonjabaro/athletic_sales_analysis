```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
Sales_2020 =pd.read_csv("../athletic_sales_analysis/Resources/athletic_sales_2020.csv")
Sales_2021 = pd.read_csv("../athletic_sales_analysis/Resources/athletic_sales_2021.csv")
Sales_2020_df = pd.DataFrame(Sales_2020)
Sales_2021_df = pd.DataFrame(Sales_2021)
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
#Because my data has the same columns I can stack the data using concat method which will add the new info below the first set of data. 
# I specify to add this on the row/horizontal axis and inner
combined_data_rows = pd.concat([Sales_2020_df,Sales_2021_df], axis="rows", join="inner")
# we reset the index with the new addtion of data using the reset_index method
combined_data_rows=combined_data_rows.reset_index()
combined_data_rows.head(25)
```

```python
# Check if any values are null.
#After importing the numpy library I use the isnull method to find nulls and the sum method to add the totals.  
nullvalues = combined_data_rows.isnull().sum()
nullvalues
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_data_rows['invoice_date'] = pd.to_datetime(combined_data_rows['invoice_date'])
combined_data_rows
```

```python
# Confirm that the "invoice_date" data type has been changed.
data_types = combined_data_rows.dtypes
print("Data types by column:")
print(data_types)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_athletic_footwear=combined_data_rows.loc[(combined_data_rows["product"]== "Women's Athletic Footwear")]
womens_athletic_footwear.head()
```

#### Using `pivot_table`

```python
#The instructions in the BCS indicate we can choose to use either the groupby or a pivot table function to review the data.  
# I chose to use the pivot table method throughout.
```

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
womens_athfootwear_retailer = pd.pivot_table(womens_athletic_footwear,
                                   values= "units_sold",
                                   index=["retailer","region","state","city"],
                                   aggfunc="sum")

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_athfootwear_retailer = womens_athfootwear_retailer.rename(columns={"units_sold":"Womens_Footwear_Units_Sold"})
# Show the top 5 results.
womens_athfootwear_retailer.sort_values(by=["Womens_Footwear_Units_Sold"],ascending=False).head()
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
womens_athfootwear_sales_days = pd.pivot_table(womens_athletic_footwear,
                                   values= "total_sales",
                                   index=["invoice_date"],
                                   aggfunc="sum")


# Optional: Rename the "total_sales" column to "Total Sales"
womens_athfootwear_sales_days =womens_athfootwear_sales_days.rename(columns={"total_sales":"Total Sales"})

# Show the table.
womens_athfootwear_sales_days.sort_values(by=["Total Sales"],ascending=False).head(10)
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
womens_daily_resample = womens_athfootwear_sales_days.resample("D").sum()

womens_daily_resample



```

```python
# Sort the resampled pivot table in ascending order on "Total Sales".
womens_daily_resample = womens_daily_resample.sort_values(by=["Total Sales"],ascending=False)
womens_daily_resample.head(10)
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
womens_weekly_resample = womens_athfootwear_sales_days.resample("W").sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
womens_weekly_resample=womens_weekly_resample.sort_values(by=["Total Sales"],ascending=False)
womens_weekly_resample.head(10)
```
